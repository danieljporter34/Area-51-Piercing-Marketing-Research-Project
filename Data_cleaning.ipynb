```python
import pandas as pd
import numpy as np
```

```python
raw_data = pd.read_excel('./Data/RawDataEditedTitles.xlsx')
raw_data.head()
```

```python
data = raw_data.copy().iloc[1:, :]
data['Drop'] = np.zeros(len(data))
data.head()
```

```python
for i in range(len(data)):
    if data.iloc[i, 0] < pd.Timestamp('2025-05-12 10:00:00'):
        data.at[data.index[i], 'Drop'] = 1  # Faster for single-cell updates

data.head()
```

```python
data['Familiar'] = data['Familiar'].replace('AnswerDisplayOrder', '')
data.head()
```

```python
df = data[['Familiar']]

# Convert to string, split digits, then one-hot encode
df_encoded = (
    df['Familiar']
    .astype(str)
    .str.split('', expand=True)  # Split into individual digits
    .stack()                     # Convert to long format
    .str.get_dummies()           # One-hot encode
    .groupby(level=0).max()      # Combine rows back
    .add_prefix('choice_')       # Rename columns (e.g., 'choice_6')
)

# Join back to original DataFrame
df = df.join(df_encoded)
df
keeper_cols = df[['choice_6', 'choice_7', 'choice_8']]
keeper_cols = keeper_cols.rename(columns={'choice_6':'FamiliarHP', 'choice_7':'FamiliarA51P', 'choice_8':'FamiliarDHP'})
keeper_cols
```

```python
data = data.join(keeper_cols)
data.head()
```

```python
data.to_excel('./Data/cleaned_data.xlsx', index=False)
```
